## Applying Functions to data

```python
df_1.head(10)
```

```python
# Student age <= 23
# Adult age <= 50 and age > 23
# Old age > 50

def compute_age_group(age):
    if age <= 23:
        return 'Student'
    elif (age > 23) and (age < 50):
        return 'Adult'
    else:
        return 'Aged'    
```

```python
df_1['Category'] = df_1['age'].apply(compute_age_group)
```
