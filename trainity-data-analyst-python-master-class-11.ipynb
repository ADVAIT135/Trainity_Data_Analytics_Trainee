# Visualizing Data using Matplotlib 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
voltage = np.linspace(0, 100, 50)
resistance = 30
current_1 = voltage / resistance
```

```python
voltage
```

```python
current_1
```

```python
plt.plot(voltage, current_1, 'r')
```

```python
plt.plot(voltage, current_1, 'r--')
```

```python
plt.plot(voltage, current_1, 'r+')
```

```python
plt.plot(voltage, current_1, 'b-')
```

```python
plt.plot(voltage, current_1, 'v')
```

```python
plt.plot(voltage, current_1, 'o')
```

```python
plt.plot(voltage, current_1, '*')
```

```python
plt.plot(voltage, current_1, 'r^')
```

```python
plt.plot(voltage, current_1, 'r^')
plt.show()
```

```python
current_2 = voltage / resistance ** 2
```

```python
plt.plot(voltage, current_1, 'r-', label='V = I * R')
plt.plot(voltage, current_2, 'k--', label = 'V = I * R^2')
plt.legend()
plt.show()
```

```python
x = np.linspace(1, 10, 100)
y = np.log(x)

plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.plot(x,y, label = 'chart 1')
plt.title("y = log(x)")
#plt.label("X")
#plt.ylabel("log(x)")
plt.legend()

plt.subplot(1,2,2)
plt.plot(x, np.exp(y), label = 'chart 2')
plt.title("y = x")
#plt.xlabel("X")
#plt.ylabel("Y")
plt.legend()

plt.show()
```

```python
x = np.linspace(1, 10, 100)
y = np.log(x)

plt.figure(figsize=(10,10))

plt.subplot(2,2,1)
plt.plot(x,y, label = 'chart 1')
plt.title("y = log(x)")
#plt.label("X")
#plt.ylabel("log(x)")
plt.legend()

plt.subplot(2,2,2)
plt.plot(x, np.exp(y), label = 'chart 2')
plt.title("y = x")
#plt.xlabel("X")
#plt.ylabel("Y")
plt.legend()

plt.subplot(2,2,(3,4))
plt.plot(x, np.sqrt(x), label = 'chart 3')
plt.title("y = x^(1/2)")
#plt.xlabel("X")
#plt.ylabel("Y")
plt.legend()


plt.tight_layout()
plt.show()
```

### Box Plot

```python
df.head()
```

```python
plt.figure(figsize = (5,5))
plt.boxplot(df['expenses'])
plt.title('Expenses Boxplot')
plt.show()
```

```python
plt.figure(figsize = (5,5))
plt.boxplot(df['bmi'])
plt.title('BMI Boxplot')
plt.show()
```

### Histogram

```python
plt.figure(figsize = (10,5))

plt.subplot(1,2,1)
plt.hist(df['expenses'], bins = 100)
plt.title('Expenses Histogram plot')

plt.subplot(1,2,2)
plt.hist(df['bmi'],  bins = 50)
plt.title('BMI Histogram plot')

plt.show()
```

### Scatterplot

```python
plt.figure(figsize = (10,5))

plt.scatter(x = df['bmi'], y = df['expenses'], c='orange')

plt.show()
```
