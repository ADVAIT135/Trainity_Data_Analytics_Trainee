# Master Class - 5 Assignment --> Array Adventure

Task:
You embark on an exciting array adventure where you'll encounter various
challenges related to NumPy arrays and their operations. Your task is to write a
Python program that allows you to overcome these challenges and progress in
your array adventure.

Requirements:
1. Create a function to display the available array challenges to the player.
2. Implement a function to create and display a NumPy array based on user input.
3. Create a function to perform basic operations on arrays, such as finding the
sum, mean, maximum, or minimum value.
4. Implement a function to reshape an array and display the reshaped version.
5. Create a function to perform element-wise mathematical operations on arrays,
such as addition, subtraction, multiplication, or division.
6. Implement a function to check if two arrays are equal or not.
7. Provide feedback and rewards to the player for completing each challenge.

```python
import numpy as np

def display_challenges():
    print("Available Challenges:")
    print("1. Create and Display a NumPy Array")
    print("2. Basic Operations on Arrays (Sum, Mean, Max, Min)")
    print("3. Reshape an Array")
    print("4. Element-wise Mathematical Operations on Arrays")
    print("5. Check if Two Arrays are Equal")

def create_and_display_array():
    size = int(input("Enter the size of the array: "))
    elements = input(f"Enter {size} elements separated by space: ").split()
    array = np.array([int(ele) for ele in elements])
    print("Created Array:\n", array)
    return array

def basic_operations(array):
    print("Sum of array:", np.sum(array))
    print("Mean of array:", np.mean(array))
    print("Maximum value in array:", np.max(array))
    print("Minimum value in array:", np.min(array))

def reshape_array(array):
    new_shape = tuple(map(int, input("Enter the new shape (e.g., 2 3 for a 2x3 array): ").split()))
    reshaped_array = array.reshape(new_shape)
    print("Reshaped Array:\n", reshaped_array)
    return reshaped_array

def element_wise_operations(array1, array2):
    print("Element-wise Addition:\n", np.add(array1, array2))
    print("Element-wise Subtraction:\n", np.subtract(array1, array2))
    print("Element-wise Multiplication:\n", np.multiply(array1, array2))
    print("Element-wise Division:\n", np.divide(array1, array2))

def check_arrays_equal(array1, array2):
    are_equal = np.array_equal(array1, array2)
    print("Arrays are equal." if are_equal else "Arrays are not equal.")
    return are_equal

def array_adventure():
    print("Welcome to the NumPy Array Adventure!")
    while True:
        display_challenges()
        choice = input("Choose a challenge (1-5) or 'q' to quit: ")
        
        if choice == '1':
            array = create_and_display_array()
            print("Challenge 1 Completed! Well done!")
        
        elif choice == '2':
            array = create_and_display_array()
            basic_operations(array)
            print("Challenge 2 Completed! Great job!")
        
        elif choice == '3':
            array = create_and_display_array()
            reshaped_array = reshape_array(array)
            print("Challenge 3 Completed! Nicely done!")
        
        elif choice == '4':
            print("Array 1:")
            array1 = create_and_display_array()
            print("Array 2:")
            array2 = create_and_display_array()
            element_wise_operations(array1, array2)
            print("Challenge 4 Completed! Excellent!")
        
        elif choice == '5':
            print("Array 1:")
            array1 = create_and_display_array()
            print("Array 2:")
            array2 = create_and_display_array()
            check_arrays_equal(array1, array2)
            print("Challenge 5 Completed! Fantastic!")
        
        elif choice.lower() == 'q':
            print("Thank you for playing the NumPy Array Adventure!")
            break
        
        else:
            print("Invalid choice. Please choose a valid challenge.")
```

```python
array_adventure()
```
