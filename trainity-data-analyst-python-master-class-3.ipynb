```python
## Write a program to compute value of 10!
def fact():
    num = int(input("Enter the Whole number whose value you need to find: "))
    fact = 1
    if num == 1:
        print("The factorial of the given number is 1")
    else:
        for i in range(1, num+1):
            fact = fact * i
    print("The factorial of the entered number",num,"is",fact)    
```

### 1. Dictionary Quests

You are on a quest to collect various items in different locations. Your task is to
create a dictionary to keep track of the items you find in each location. Write a
Python program that allows you to add items to the dictionary, remove items, and
check if a specific item exists in a location.
Requirements:
1. Create an empty dictionary to represent the player's inventory.
2. Prompt the user to enter a location and an item found in that location.
3. Add the item to the inventory dictionary with the location as the key and the item
as the value.
4. Allow the user to remove an item from a specific location.
5. Implement a function to check if a specific item exists in any of the locations in
the inventory.
6. Display the current inventory after each operation.

```python
def add_item(inventory, location, item):
    if location in inventory:
        inventory[location].append(item)
    else:
        inventory[location] = [item]
        
def remove_item(inventory, location, item):
    if location in inventory and item in inventory[location]:
        inventory[location].remove(item)
        if not inventory[location]:
            del inventory[location]


def check_item(inventory, item):
    for location, items in inventory.items():
        if item in items:
            return True
    return False

def display_inventory(inventory):
    if not inventory:
        print("Inventory is empty.")
    else:
        for location, items in inventory.items():
            print(f"{location}: {', '.join(items)}")
```

```python
def main():
    inventory = {}
    
    while True:
        print("\nOptions:")
        print("1. Add an item")
        print("2. Remove an item")
        print("3. Check if an item exists")
        print("4. Display inventory")
        print("5. Exit")
        
        choice = input("Choose an option: ")
        
        if choice == '1':
            location = input("Enter the location: ")
            item = input("Enter the item found: ")
            add_item(inventory, location, item)
        elif choice == '2':
            location = input("Enter the location: ")
            item = input("Enter the item to remove: ")
            remove_item(inventory, location, item)
        elif choice == '3':
            item = input("Enter the item to check: ")
            if check_item(inventory, item):
                print(f"{item} exists in the inventory.")
            else:
                print(f"{item} does not exist in the inventory.")
        elif choice == '4':
            display_inventory(inventory)
        elif choice == '5':
            break
        else:
            print("Invalid choice. Please choose a valid option.")
```

```python
main()
```

## 2. Set Collector

You are a collector of unique items. Your task is to create sets of items based on
their categories and perform operations on those sets. Write a Python program
that allows you to create sets, combine sets, find common elements, and check if
one set is a proper subset of another set.
 Requirements:
1. Prompt the user to enter items and their categories.
2. Create sets of items based on their categories.
3. Allow the user to combine two sets into a single set.
4. Implement a function to find and display the common elements between two
sets.
5. Implement a function to check if one set is a proper subset of another set.
6. Display the sets and the results of the operations after each step.

```python
def add_item_to_category(categories, category, item):
    if category in categories:
        categories[category].add(item)
    else:
        categories[category] = {item}

def combine_sets(categories, category1, category2, new_category):
    if category1 in categories and category2 in categories:
        categories[new_category] = categories[category1].union(categories[category2])
    else:
        print("One or both categories do not exist.")

def common_elements(categories, category1, category2):
    if category1 in categories and category2 in categories:
        return categories[category1].intersection(categories[category2])
    else:
        print("One or both categories do not exist.")
        return set()

def is_proper_subset(categories, subset_category, superset_category):
    if subset_category in categories and superset_category in categories:
        return categories[subset_category] < categories[superset_category]
    else:
        print("One or both categories do not exist.")
        return False

def display_categories(categories):
    if not categories:
        print("No categories available.")
    else:
        for category, items in categories.items():
            print(f"{category}: {', '.join(items)}")

def set_main():
    categories = {}
    
    while True:
        print("\nOptions:")
        print("1. Add an item to a category")
        print("2. Combine two sets")
        print("3. Find common elements between two sets")
        print("4. Check if one set is a proper subset of another")
        print("5. Display categories")
        print("6. Exit")
        
        choice = input("Choose an option: ")
        
        if choice == '1':
            category = input("Enter the category: ")
            item = input("Enter the item: ")
            add_item_to_category(categories, category, item)
        elif choice == '2':
            category1 = input("Enter the first category to combine: ")
            category2 = input("Enter the second category to combine: ")
            new_category = input("Enter the name of the new combined category: ")
            combine_sets(categories, category1, category2, new_category)
        elif choice == '3':
            category1 = input("Enter the first category: ")
            category2 = input("Enter the second category: ")
            common = common_elements(categories, category1, category2)
            if common:
                print(f"Common elements between {category1} and {category2}: {', '.join(common)}")
        elif choice == '4':
            subset_category = input("Enter the subset category: ")
            superset_category = input("Enter the superset category: ")
            if is_proper_subset(categories, subset_category, superset_category):
                print(f"{subset_category} is a proper subset of {superset_category}.")
            else:
                print(f"{subset_category} is not a proper subset of {superset_category}.")
        elif choice == '5':
            display_categories(categories)
        elif choice == '6':
            break
        else:
            print("Invalid choice. Please choose a valid option.")

```

```python
set_main()
```

## 3. Number Challenge 

You have encountered a number challenge! Your task is to solve different
number-related puzzles using control flow statements. Write a Python program
that prompts the user to enter a number and perform operations to determine
properties of the number.
 Requirements:
1. Prompt the user to enter a number.
2. Check if the number is positive, negative, or zero, and display the result.
3. Check if the number is even or odd, and display the result.
4. Prompt the user to enter a divisor and check if the number is a multiple of the
divisor, displaying the result.
5. Provide feedback to the user based on their answers and encourage them to
solve more number challenges.

```python
def pos_neg(num):
    if num < 0:
        print("The entered number",num,"is a negative number")
    elif num > 0:
        print("The entered number",num,"is a positive number")
    elif num == 0:
        print("The entered number",num,"is 0")
        
def even_odd(num):
    if (num%2 == 0):
        print("The entered number",num,"is an Even number")
    elif (num%2 != 0):
        print("The entered number",num,"is an Odd number")
            
def mul_div(num):
    div = int(input("Enter the divisor(whole number): "))
    if (num%div == 0):
        print("The entered number",num,"is a multiple of entered divisor",div)
    else:
        print("The entered number",num,"is a not a multiple of entered divisor",div)   
```

```python
def num_main():
    num = int(input("Enter any whole number(number without decimal points): "))
    pos_neg(num)
    even_odd(num)
    mul_div(num)
```

```python
num_main()
```

```python
num_main()
```
