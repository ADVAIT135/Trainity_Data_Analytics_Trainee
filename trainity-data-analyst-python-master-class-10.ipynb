```python
import pandas as pd
```

```python
df = pd.read_csv('/kaggle/working/insurance-dataset-regression-analysis.zip')
```

```python
df['family_count'] = df['children'] + 1
df['Avg_expense_per_head'] = df['expenses'] / df['family_count']
```

```python
df.head()
```

```python
df_avg_premium_age = df.groupby('age')['Avg_expense_per_head'].mean()
```

```python
df_avg_premium_age = df_avg_premium_age.to_frame().reset_index()
```

```python
df_avg_premium_age.head()
```

```python
df_avg_premium_age.rename(columns = {'Avg_expense_per_head':'avg_premium'}, inplace=True)
```

```python
df_avg_premium_age.head()
```

##### how = inner/outer/left/right

```python
pd.merge(df, df_avg_premium_age, on='age', how='inner')
```

## Basic Plotting with Matplotlib, Axes Labels and Titles


```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Plotting two 1-D numpy arrays
x = np.linspace(5, 100, 100)
y = np.linspace(10,1000, 100)
```

```python
ax = plt.plot(x,y)
```

```python
voltage = np.linspace(0, 100, 50)
resistance = 30

# V = IR

current = voltage / resistance
```

```python
plt.plot(voltage, current, label = 'V=I*R')
plt.legend()
plt.xlabel('Voltage')
plt.ylabel('Current')

plt.title("Ohm's Law")
plt.show()
```

```python
plt.plot(voltage, current, label = 'V=I*R')
plt.legend()
plt.xlabel('Voltage')
plt.ylabel('Current')

plt.xlim([20,60])

plt.title("Ohm's Law")
plt.show()
```

```python
plt.plot(voltage, current, label = 'V=I*R')
plt.legend()
plt.xlabel('Voltage')
plt.ylabel('Current')

plt.xlim([20,60])
plt.ylim([1,3])

plt.title("Ohm's Law")
plt.show()
```

```python
current_2 = voltage / resistance ** 2
```

```python
plt.plot(voltage, current, label = 'V = I*R')
plt.plot(voltage, current_2, label = 'V = I*R ^ 2')

plt.xlabel('Voltage')
plt.ylabel('Current')

plt.legend()

plt.title('Results of Exp 1 and 2')

plt.show()
```
