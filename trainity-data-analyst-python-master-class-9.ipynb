# Data Analysis using Pandas

```python
import pandas as pd
```

```python
!kaggle datasets download -d mahejabeenabdul/insurance-dataset-regression-analysis
```

```python
df = pd.read_csv('/kaggle/working/insurance-dataset-regression-analysis.zip')
```

```python
df.head()
```

```python
df['family_count'] = df['children'].astype(int) + 1
```

```python
df.head()
```

```python
df['Avg_expense_per_head'] = df['expenses'] / df['family_count']
```

```python
df.head()
```

```python
df.describe()
```

```python
df[['bmi','expenses']]
```

```python
df.plot(x='bmi', y='expenses', kind='scatter')
```

```python
df.plot(x='age', y='expenses', kind='scatter')
```

```python
df.plot(x='children', y='expenses', kind='scatter')
```

```python
df.plot(x='family_count', y='expenses', kind='scatter')
```

```python
df.groupby('region')[['Avg_expense_per_head']].median()
```

```python
df.groupby('region')[['Avg_expense_per_head']].median().plot(kind='bar')
```

```python
df.groupby('region')[['Avg_expense_per_head']].median().plot(kind='barh')
```
