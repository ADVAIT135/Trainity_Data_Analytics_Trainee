# Master Class 4 -- Assignment

You are on a quest to explore and gather information from various files in a
mysterious land. Your task is to create a Python program that simulates this
quest and allows you to perform different file handling operations.
Requirements:
1. Create a function to display the available file options to the player.
2. Implement a function to read the contents of a file and display them to the
player.
3. Create a function to write new information to a file.
4. Implement a function to append additional information to an existing file.
5. Create a function to delete a file from the quest.
6. Implement a function to search for specific content within a file and display the
search results.
7. Provide feedback and rewards to the player for completing each operation.

```python
import os

# Directory where the files are stored
quest_directory = "quest_files"

# Ensure the directory exists
os.makedirs(quest_directory, exist_ok=True)

# Display available file options to the player
def display_file_options():
    files = os.listdir(quest_directory)
    if files:
        print("Available files:")
        for file in files:
            print(f"- {file}")
    else:
        print("No files available.")

# Read the contents of a file and display them to the player
def read_file(filename):
    filepath = os.path.join(quest_directory, filename)
    try:
        with open(filepath, 'r') as file:
            contents = file.read()
        print(f"Contents of {filename}:\n{contents}")
        print("Reward: You gained 10 points!")
    except FileNotFoundError:
        print(f"File {filename} not found.")

# Write new information to a file
def write_file(filename, content):
    filepath = os.path.join(quest_directory, filename)
    with open(filepath, 'w') as file:
        file.write(content)
    print(f"Written to {filename}.")
    print("Reward: You gained 20 points!")

# Append additional information to an existing file
def append_file(filename, content):
    filepath = os.path.join(quest_directory, filename)
    with open(filepath, 'a') as file:
        file.write(content)
    print(f"Appended to {filename}.")
    print("Reward: You gained 15 points!")

# Delete a file from the quest
def delete_file(filename):
    filepath = os.path.join(quest_directory, filename)
    try:
        os.remove(filepath)
        print(f"File {filename} deleted.")
        print("Reward: You gained 25 points!")
    except FileNotFoundError:
        print(f"File {filename} not found.")

# Search for specific content within a file and display the search results
def search_file(filename, search_term):
    filepath = os.path.join(quest_directory, filename)
    try:
        with open(filepath, 'r') as file:
            contents = file.read()
        occurrences = contents.count(search_term)
        print(f"Found '{search_term}' {occurrences} time(s) in {filename}.")
        print("Reward: You gained 30 points!")
    except FileNotFoundError:
        print(f"File {filename} not found.")

# Main quest loop
def quest():
    while True:
        print("\n--- Quest Menu ---")
        print("1. Display file options")
        print("2. Read a file")
        print("3. Write to a file")
        print("4. Append to a file")
        print("5. Delete a file")
        print("6. Search in a file")
        print("7. Exit")

        choice = input("Choose an option: ")

        if choice == "1":
            display_file_options()
        elif choice == "2":
            filename = input("Enter the filename to read: ")
            read_file(filename)
        elif choice == "3":
            filename = input("Enter the filename to write to: ")
            content = input("Enter the content to write: ")
            write_file(filename, content)
        elif choice == "4":
            filename = input("Enter the filename to append to: ")
            content = input("Enter the content to append: ")
            append_file(filename, content)
        elif choice == "5":
            filename = input("Enter the filename to delete: ")
            delete_file(filename)
        elif choice == "6":
            filename = input("Enter the filename to search in: ")
            search_term = input("Enter the search term: ")
            search_file(filename, search_term)
        elif choice == "7":
            print("Exiting the quest. Goodbye!")
            break
        else:
            print("Invalid choice. Please try again.")
```

```python
quest()
```
