```python
user_str = input("Enter any string: ")
print("The entered string value is",user_str)
print("The length of the entered string is",len(user_str))
print("The entered string in it's uppercase is",user_str.upper())
print("The entered string contains only alphabetic characters",user_str.isalpha())
```

## 4. Write a Python program that demonstrates various list manipulation operations, such as:
## Appending elements to a list.
## Removing elements from a list.
## Accessing elements by index.
## Slicing a list to extract a sublist.
## Sorting the elements of a list in ascending order

```python
lst = []
#Adding 20 even numbers from 1 to 40 in the lst
for i in range(0,41):
    if i%2 == 0:
        #Appending Elements(in this case even numbers from 0 to 40) to list 
        lst.append(i)
print(lst), print(len(lst))
```

```python
# Let's remove 0 from the above list cause 0 is niether even nor odd
lst.remove(0)
```

```python
print(lst)
```

```python
## Let's Extract Numbers which are multiples of 8
print(lst[3],lst[7],lst[11],lst[15],lst[19])
```

```python
# Let's extract numbers which are greater than 10
print(lst[5:])
```

```python
# We have sorted the list in Descending order, since the list we generated was already sorted in Ascending order
lst.sort(reverse=True)
```

## 5. Write a Python program that prompts the user to enter two tuples and performs the following operations:
## Concatenate the two tuples and print the result.
## Check if a specific element exists in either of the tuples and print the result
## (True or False).

```python
tuple_1 = input(("Enter the elements of First Tuple seperated by commas"))
tuple_2 = input(("Enter the elements of Second Tuple seperated by commas"))

tuple_A = tuple(tuple_1.split(','))
tuple_B = tuple(tuple_2.split(','))

print("First Tuple: ",tuple_A)
print("Second Tuple: ",tuple_B)

print("Concatenation of the two tuples is",tuple_A+tuple_B)

elem = input("Enter the specific element that you want to look for either of the tuples: ")
print((elem in tuple_A)or(elem in tuple_B))
```
