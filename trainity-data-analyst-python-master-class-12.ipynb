# Data Visualization using Seaborn

```python
df = pd.read_csv('/kaggle/working/insurance-dataset-regression-analysis.zip')
```

```python
df.head()
```

```python
df_avg_age_exp = df.groupby('age')['expenses'].mean().to_frame().reset_index()
```

```python
df_avg_age_exp.head()
```

```python
sns.lineplot(x = 'age', y='expenses', data=df)
plt.title('Average Expense vs Age')
plt.show()
```

```python
sns.lineplot(x = 'age', y='expenses', data=df_avg_age_exp)
plt.title('Average Expense vs Age')
plt.show()
```

```python
df_premium = df.groupby('region')['expenses'].mean().to_frame().reset_index()
```

```python
df_premium.head()
```

```python
sns.barplot(x='region', y='expenses', data =df_premium)
plt.title('Average Premium by Region')
plt.show()
```

```python
df_premium_region_sex = df.groupby(['region','sex'])['expenses'].mean().to_frame().reset_index()
```

```python
df_premium_region_sex
```

```python
plt.figure(figsize = (10,5))
sns.barplot(x = 'region', y = 'expenses', hue = 'sex', data = df_premium_region_sex)
plt.title('Average Premium paid by Region and Sex of the Customer')
plt.show()
```

```python
sns.displot(x='expenses', data = df)
plt.show()
```

```python
sns.displot(x='expenses', data = df, kind='kde')
plt.show()
```

```python
sns.boxplot(x='region', y='bmi', data=df)
plt.show()
```

```python
sns.violinplot(x='bmi',y='region', data = df)
plt.show()
```

```python
sns.relplot(x='age',y='bmi', data = df, kind='scatter')
plt.show()
```

```python
sns.relplot(x='age',y='bmi', data = df, kind='scatter', hue='sex')
plt.show()
```

```python
sns.relplot(x='age',y='bmi', data = df, kind='scatter', hue='region')
plt.show()
```

```python
sns.relplot(x='age',y='bmi', data = df, kind='line', hue='sex')
plt.show()
```

```python
plt.figure(figsize = (10,10))
sns.relplot(x='age',y='expenses', size='bmi', data=df, hue='bmi')
plt.show()
```

```python
plt.figure(figsize = (10,10))
sns.relplot(x='age',y='expenses', size='bmi', data=df, alpha=0.5)
plt.show()
```

```python
plt.figure(figsize = (10,10))
sns.relplot(x='age',y='expenses', size='bmi', data=df,
            row='sex', col='region', alpha=0.5)
plt.show()
```

```python
sns.catplot(x = df['region'],y = df['bmi'], hue=df['region'])
plt.show()
```

```python
sns.catplot(x = 'region',y = 'bmi',data=df, kind='point')
plt.show()
```

```python
sns.catplot(x = 'region',y = 'bmi',data=df, kind='point', hue='sex')
plt.show()
```

```python
sns.catplot(x = 'region',y = 'expenses',data=df, kind='point', hue='sex')
plt.show()
```

```python
sns.catplot(x = 'region',y = 'expenses',data=df, kind='point')
plt.show()
```
